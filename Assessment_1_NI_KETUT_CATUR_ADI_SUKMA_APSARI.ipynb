mengimport library yang digunakan, seperti seaborn, pandas

```python
import seaborn as sns
import pandas as pd


```

Mengimport dataset dengan menggunakan pandas, dan menampilkan data

```python

data_set = 'https://raw.githubusercontent.com/dataenergy/Exploratory-data-analysis-of-climate-change-and-natural-disasters/master/datasets/number-of-natural-disaster-events.csv'
data = pd.read_csv(data_set)
data.head()
```

menampilkan barplot horizontal dengan seaborn

```python
sns.barplot(x="Year", y="Entity", ci = None , data=data )
```

Menampilkan barplot vertikal dengan seaborn

```python
sns.barplot(y="Year", x="Entity", ci = None , data=data )
```

menampilkan displot

```python
sns.displot (data=data, x="Number of reported natural disasters (reported disasters)", kde=True)
```
